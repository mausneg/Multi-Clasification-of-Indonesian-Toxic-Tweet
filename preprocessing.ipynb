```python
import numpy as np
import pandas as pd
import re
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
```

```python
alay_dict = pd.read_csv('Dataset/new_kamusalay.csv', encoding='latin-1', header=None)
alay_dict = alay_dict.rename(columns={0: 'original', 
                                      1: 'replacement'})
```

```python
id_stopword_dict = pd.read_csv('Dataset/stopwordbahasa.csv', header=None)
id_stopword_dict = id_stopword_dict.rename(columns={0: 'stopword'})
```

```python
factory = StemmerFactory()
stemmer = factory.create_stemmer()
```

```python
def lowercase(text):
    return text.lower()
```

```python
def remove_unnecessary_char(text):
    text = re.sub('\n',' ',text)
    text = re.sub('rt',' ',text)
    text = re.sub('user',' ',text)
    text = re.sub('((www\.[^\s]+)|(https?://[^\s]+)|(http?://[^\s]+))',' ',text)
    text = re.sub('  +', ' ', text)
    text = re.sub(r'\\x', ' ', text)
    text = re.sub(r'x[a-fA-F0-9]{2}', ' ', text)
    return text
```

```python
def remove_nonaplhanumeric(text):
    text = re.sub('[^0-9a-zA-Z]+', ' ', text) 
    return text
```

```python
data['Tweet'] = data['Tweet'].apply(lowercase)
data['Tweet'] = data['Tweet'].apply(remove_nonaplhanumeric)
data['Tweet'] = data['Tweet'].apply(remove_unnecessary_char)
```

```python
data.head(10)
```

```python
alay_dict_map = dict(zip(alay_dict['original'], alay_dict['replacement']))
def normalize_alay(text):
    return ' '.join([alay_dict_map[word] if word in alay_dict_map else word for word in text.split(' ')])
```

```python
def remove_stopword(text):
    text = ' '.join(['' if word in id_stopword_dict.stopword.values else word for word in text.split(' ')])
    text = re.sub('  +', ' ', text)
    text = text.strip()
    return text
```

```python
def stemming(text):
    return stemmer.stem(text)
```

```python
data['Tweet'] = data['Tweet'].apply(normalize_alay)
data['Tweet'] = data['Tweet'].apply(remove_stopword)
data['Tweet'] = data['Tweet'].apply(stemming)
```
